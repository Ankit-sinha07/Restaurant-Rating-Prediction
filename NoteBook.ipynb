```python
x = df[["Average Cost for two", "Has Table booking","Has Online delivery", "Price range"]]
```

```python
y = df[["Aggregate rating"]]
```

```python
x
```

```python
from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()
```

```python
x["Has Table booking"] = labelencoder.fit_transform(x["Has Table booking"])
```

```python
x
```

```python
x["Has Online delivery"] = labelencoder.fit_transform(x["Has Online delivery"])
```

```python
labelencoder.classes_
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size=0.2) #20% of the data is going to be in the test set and the 80% will be in the training set
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
```

```python
#Function for getting the model results
def modelresult(predictions):
    print("Mean absolute error on model is {}".format(mean_absolute_error(Y_test, predictions)))
    print("Root mean squared error on model is {}".format(mean_squared_error(Y_test,predictions)))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
predictionsfromlr = lr.predict(X_test)
```

```python
modelresult(predictionsfromlr)
```

```python
predictionsfromlr
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
svrmodel = SVR()
```

```python
# Define the parameter grid with the correct lowercase parameter name
param_gridsvr = {
    'C': [0.1,0.5],
    'epsilon': [0.01, 0.1, 1],
    'kernel': ['linear', 'poly'],"degree" : [ 2 ] # Use 'kernel' instead of 'Kernel'
}
# Instantiate and run GridSearchCV
gridsvr  = GridSearchCV(svrmodel, param_gridsvr)
gridsvr.fit(X_train, Y_train)
```

```python
predsgridsvr = gridsvr.predict(X_test)
```

```python
modelresult(predsgridsvr)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
param_grid={
    'max_depth':[10,20],
    'min_samples_leaf':[1,2],
    'min_samples_split':[2,5]
}
```

```python
tree_model = DecisionTreeRegressor()
```

```python
#Imagine grid_tree is an instance of GridSearchCV
grid_tree = GridSearchCV(tree_model, param_grid)

# Fit it to your training data
grid_tree.fit(X_train, Y_train)
```

```python
treepredictions = grid_tree.predict(X_test)
```

```python
modelresult(treepredictions)
```
